```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
image=cv2.imread("hunk.jpg")
plt.figure()
plt.axis("off")
plt.imshow(image)
image.shape
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)#This is the real image..
plt.figure()
plt.axis("off")
plt.imshow(image)
image.shape
```

```python
image = image.reshape((image.shape[0] * image.shape[1], 3))
print image
print image.shape
```

```python
from sklearn.cluster import KMeans
K = 3
clf = KMeans(n_clusters=K)
clf.fit(image)
```

```python
label = [0, 0, 0]

for x in clf.labels_:
    label[x] += 1

label_sum = image.shape[0]

label = [float(x) / label_sum for x in label]
print label
```

```python

import webcolors

centroids = [x.astype("uint8").tolist() for x in clf.cluster_centers_]
clr = [webcolors.rgb_to_hex(c) for c in centroids]

plt.pie(label, colors=clr, shadow=True, startangle=90,  autopct="%0.0f%%") 
plt.axis('equal')
plt.show()
```
